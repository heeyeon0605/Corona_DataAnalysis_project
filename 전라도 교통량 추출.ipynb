```python
import pandas as pd
import os
```

```python
path = 'C:/Users/위나현/Desktop/프로젝트/고속도로 톨게이트 입출구 교통량 데이터/'
file_list = os.listdir(path)
file_list_분기 = [file for file in file_list if file.endswith('분기.csv')]
file_list_분기
```

```python
df = pd.DataFrame()
for i in file_list_분기:
    data = pd.read_csv(path + i, encoding = 'cp949')
    df = pd.concat([df,data])
df
```

```python
df컬럼삭제 = df.drop(columns = ['입출구구분코드', 'TCS하이패스구분코드', '고속도로운영기관구분코드', '고속도로운영기관명', 
                            '영업형태구분코드', '영업형태명'])
df컬럼삭제
```

```python
# 전라도의 외부방문자 상위 15개 시에 해당하는 영업소
전라 = ['동전주', '서전주', '전주', '군산', '동군산', '남순천', '동순천', '서순천', '순천', '순천만', '광주', '동광주', 
          '북광주', '동광산', '서광산', '무안', '무안공항', '북무안', '익산', '완주', '목포', '나주', '담양', '주암송광사', 
          '승주', '황전', '일로', '삼례', '소양', '상관', '문평', '창평']
```

```python
df전라 = df컬럼삭제[df컬럼삭제['영업소명'].isin(전라)]
df전라
```

```python
# 입구 교통량만
df입구 = df전라[df전라.입출구명 == '입구']
df입구
```

```python
df입구.drop(columns = '입출구명', inplace = True)
df입구
```

```python
# TCS하이패스 구분없이 교통량 합산
df합 = df입구.groupby(['집계일자', '영업소명', '영업소코드'], as_index = False).sum()
df합
```

```python
# 지역별 영업소
Jeonju = ['동전주', '서전주', '전주']
Gunsan = ['군산', '동군산']
Suncheon = ['남순천', '동순천', '서순천', '순천', '순천만', '주암송광사', '승주', '황전']
Gwangju = ['광주', '동광주', '북광주', '동광산', '서광산']
Muan = ['무안', '무안공항', '북무안', '일로']
Iksan = ['익산', '삼례']
Wanju = ['완주', '소양', '상관']
#목포
Naju = ['나주', '문평']
Damyang = ['담양', '창평']
```

```python
지역 = []
for i in df합.영업소명:
    if i in Jeonju:
        지역.append('Jeonju')
    elif i in Gunsan:
        지역.append('Gunsan')
    elif i in Suncheon:
        지역.append('Suncheon')
    elif i in Gwangju:
        지역.append('Gwangju')
    elif i in Muan:
        지역.append('Muan')
    elif i in Iksan:
        지역.append('Iksan')
    elif i in Wanju:
        지역.append('Wanju')
    elif i == '목포':
        지역.append('Mokpo')
    elif i in Naju:
        지역.append('Naju')
    elif i in Damyang:
        지역.append('Damyang')
    else:
        print(i)
len(지역)
```

```python
df합['province'] = 지역
df합
```

```python
# 열 이름 영어로 바꿔주기
df합.columns = ['date', 'tollgate', 'code', '1jong', '2jong', '3jong', '4jong', '5jong', '6jong', 'total', 'province']
df합
```

```python
df합.to_csv('전라_교통량.csv', encoding = 'cp949')
```
