```python
# 강수량 0.0으로 표시된 값은 반올림 결과이기 때문에 0.05로 수정
df기온['강수량'] = df강수량['강수량(mm)'].replace(0, 0.05)
df기온[df기온['강수량'] == 0.05]
```

```python
# 결측값이 없는 강수량은 0으로 수정
df기온['강수량'] = df기온['강수량'].fillna(0)
df기온.info()
```
