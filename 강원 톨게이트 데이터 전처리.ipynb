## 모든 분기별 데이터 읽어오기

```python
import pandas as pd
from pandas import DataFrame, Series

file_list_years = ['2021_1.csv', '2020_1.csv', '2020_2.csv', '2020_3.csv', '2020_4.csv']
```

```python
# data_temp1 = pd.read_csv('tollgate_gangwon/'+file_list_years[0], encoding = 'cp949')
```

```python
data_raw = pd.DataFrame()

for i in file_list_years :
    data_temp = pd.read_csv('tollgate_gangwon/'+i, encoding = 'cp949')
    
    data_raw = pd.concat([data_raw, data_temp], axis=0)
    
data_raw
```

## 강원도 데이터만 추출

```python
Gangwon = ['춘천', '홍천', '동홍천', '원주영업소', '북원주', '남원주', '신림', '문막', '대관령', '남강릉', '강릉',
           '옥계영업소', '북강릉', '속초', '동홍천', '내촌', '홍천', '면온', '평창', '속사', '진부', '동해', '인제',
           '남양양영업소', '남양양', '삼척', '근덕', '북양양', '서양양영업소', '양양', '하조대', '망상']

data_gangwon = data_raw[data_raw['영업소명'].isin(Gangwon)]

data_gangwon
```

```python
#data_raw[data_raw["영업소명"] == "남양양영업소"]
```

## 필요없는 column 삭제

```python
data_gangwon = data_gangwon.drop(columns=['영업형태구분코드', 'TCS하이패스명', '고속도로운영기관명', '고속도로운영기관구분코드', '입출구명'
                          , '고속도로운영기관구분코드', '영업형태구분코드', '영업형태명', '4종교통량', '5종교통량'])

data_new = data_gangwon.reset_index()

data_new
```

## 집계일자와 영업소명으로 정렬

```python
data_new.sort_values(by=['집계일자','영업소명'], inplace=True)
```

```python
data_temp = data_new[data_new['입출구구분코드']==0]
```

```python
data_entrance = data_temp.drop(columns=['입출구구분코드'])
```

```python
data_entrance = data_entrance.reset_index()

data_entrance.drop(columns=['index', 'level_0'], inplace=True)
```

```python
data_entrance
```

## TCS와 하이패스 row 합치기

```python
data_TCS = data_entrance.groupby(['집계일자', '영업소명', '영업소코드'], as_index=False).sum()

data_TCS.drop(columns=['TCS하이패스구분코드'], inplace=True)

data_TCS
```

## 지역 이름 제외하고 영어로 바꾸기

```python
data_TCS.columns = ['date', 'tollgate', 'code', '1jong', '2jong', '3jong', '6jong', 'total']

data_TCS
```
