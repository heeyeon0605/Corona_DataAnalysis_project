## 강원 전지역 데이터 합치기

```python
import pandas as pd
from pandas import DataFrame, Series

file_list_rain = ['북춘천.csv', '춘천.csv', '원주.csv', '대관령.csv', '강릉.csv', '북강릉.csv', 
             '속초.csv', '홍천.csv', '동해.csv', '인제.csv', '삼척.csv', '정선.csv', '철원.csv', '태백.csv']

df_rain = pd.DataFrame()

for i in file_list_rain :
    data = pd.read_csv('강수량 지역별/'+i, encoding = 'cp949', skiprows = 7)
    
    if i[2] == '.' :
        data['지역'] = i[:2]
    else :
        data['지역'] = i[:3]
    df_rain = pd.concat([df_rain, data])
    
df_rain = df_rain.reset_index(drop = True)
```

```python
df_rain # 강원도 강수량 데이터
```

```python
df_tem = pd.DataFrame()

for i in file_list_rain :
    data = pd.read_csv('기온 지역별/'+i, encoding = 'cp949', skiprows = 7)
    
    if i[2] == '.' :
        data['지역'] = i[:2]
    else :
        data['지역'] = i[:3]
    df_tem = pd.concat([df_tem, data])
    
df_tem = df_tem.reset_index(drop = True)
```
